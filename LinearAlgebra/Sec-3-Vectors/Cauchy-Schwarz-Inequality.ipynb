```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create 3 vectors, one of which is dependen on another

a = np.random.randn(5)      # linearly independent
b = np.random.randn(5)      # linearly independent
c = np.random.randn(1) * a  # linearly dependent on a

# Compute their dot products
aTb = np.dot(a, b)
aTc = np.dot(a, c)

# demonstrate the (in)equalities
print(f'{np.abs(aTb)}, {np.linalg.norm(a) * np.linalg.norm(b)}')
print(f'{np.abs(aTc)}, {np.linalg.norm(a) * np.linalg.norm(c)}')
```

```python
# Three vectors
a = np.array([1,-2])
b = np.array([2, 3])
c = np.array([0, 2])

#draw them
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot([0, a[0]], [0, a[1]], 'r')
ax.plot([0, b[0]], [0, b[1]], 'g')
ax.plot([0, c[0]], [0, c[1]], 'b')

# Make the plot look nicer
ax.plot([0,0], [0,0], [-4,4], 'k--')
ax.plot([0,0], [-4,4], [0,0], 'k--')
ax.plot([-4,4], [0,0], [0,0], 'k--')


plt.show()

print(f'Dot product a and b : {np.dot(a, b)}')
print(f'Dot product a and c : {np.dot(a, c)}')
print(f'Dot product b and c : {np.dot(b, c)}')
```
