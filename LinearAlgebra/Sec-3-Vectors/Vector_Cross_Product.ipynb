Vector Cross Product

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as mplot3d
```

```python
# Create vectors
v1 = [-3, 2, 5]
v2 = [4, -3, 0]

# Python's cross-product fuction
v3a = np.cross (v1, v2)

print(v3a)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw a plane defined by span of v1 and v2
# Essentially, we are creating a grid, drawing a line on the grid z1 (red line)
xx, yy = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))
z1 = (-v3a[0]*xx - v3a[1]*yy)/v3a[2]
ax.plot_surface(xx, yy, z1)

## Plot the two vectors
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'k')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'k')
ax.plot([0, v3a[0]], [0, v3a[1]], [0, v3a[2]], 'r')

ax.view_init(azim=150, elev=45)
plt.show()
```
