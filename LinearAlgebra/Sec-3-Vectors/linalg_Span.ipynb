Span of a set of vectors

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as mplot3d
```

```python
# Set S
S1 = np.array([1,1,0])
S2 = np.array([1,7,0])

# vectors
v = np.array([1,2,0])
w = np.array([3,2,1])

# draw vectors
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot([0, S1[0]], [0,S1[1]], [0, S1[2]], 'r', linewidth=3)
ax.plot([0, S2[0]], [0,S2[1]], [0, S2[2]], 'r', linewidth=3)

ax.plot([0, v[0]], [0, v[1]], [0, v[2]], 'g', linewidth=3)
ax.plot([0, w[0]], [0, w[1]], [0, w[2]], 'b', linewidth=3)

# Make plot look a bit nicer
#ax.plot([0,0], [0,0], [-4,4], 'k--')
#ax.plot([0,0], [-4,4], [0,0], 'k--')
#ax.plot([-4,4], [0,0], [0,0], 'k--')

# Draw plane
xx, yy = np.meshgrid(range(-15,16), range(-15,16))
cp = np.cross(S1, S2)
z1 = (-cp[0] * xx - cp[1]*yy)/cp[2] 
ax.plot_surface(xx,yy,z1)
plt.show()

```
