```python
import numpy as np
```

```python
# topic: the linearity of trace

# topic-1: determine the relationship between tr(A) + tr(B) and tr(A+B)
# topic-2: determine the relationship between tr(l * A) and l * tr(A)
```

```python
# topic-1 : Determine the relation between tr(A) + tr(B)

N = 3

A = np.random.randn(N, N)
B = np.random.randn(N, N)

trA = np.trace(A)
trB = np.trace(B)

res_LHS = trA + trB

res_RHS = np.trace(A + B)

print(res_LHS, res_RHS)

# Trace is cummutative property
```

```python
# topic 2: Determine the distributive relation

l = 0.3

res_LHS = np.trace(l * A)
res_RHS = l * np.trace(A)

print(res_LHS), print(' '), print(res_RHS)

# trace has a distributive property
```
