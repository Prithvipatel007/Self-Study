```python
import numpy as np
```

# A Zoo of Matrices

```python
# Square vs Rectangular
S = np.random.randn(5,5)
R = np.random.randn(5,2)
print(S), print(' ')
print(R)
```

```python
# Identity
I = np.eye(3,3)
print(I)
```

```python
# zeros
# Z = np.zeros((4,4))
Z = np.zeros((4,4), dtype=complex)
print(Z)
```

```python
# Diagonal
D = np.diag([1,2,3,5,2])
print(D)
```

```python
# Triangular Matrix from Full matrix
S = np.random.randn(5,5)
U = np.triu(S)
L = np.tril(S)
print(U), print(' ')
print(L)
```

```python
# Concatenate Matrices
A = np.random.randn(3,2)
B = np.random.randn(3,4)
C = np.concatenate((A,B), axis=1)
print(C), print(' ')
```
