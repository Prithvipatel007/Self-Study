```python
import numpy as np
```

```python
A = np.reshape(np.arange(1,13), (3,4), 'F') # F = Columns, C = row

r = [10, 20, 30, 40]
c = [100, 200, 300]

print(A), print(' ')
print(r), print(' ')
print(c), print(' ')
```

```python
# broadcast on the rows
print(A + r), print(' ')

# broadcast on the columns
print( A + np.reshape(c, (len(c),1) ) ), print(' ') # only works for explicit column vectors
```
