```python
import numpy as np
```

```python
M = np.round(6* np.random.randn(4,4))

print(M), print(' ')
d = np.diag(M)

# notice the two ways of using the diag function
d = np.diag(M)
D = np.diag(d)

print(d)
print(D)

# trace as sum of diagonals
tr = np.trace(M)
tr2 = sum(np.diag(M))
print(' '), print(tr, tr2)
```
