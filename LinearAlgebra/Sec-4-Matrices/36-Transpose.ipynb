```python
import numpy as np
```

```python
M = np.array([ [1,2,3], [2,3,4] ])

print(M), print(' ')
print(M.T), print(' ')
print(M.T.T), print(' ')

# can also use the function transpose
print(np.transpose(M))
```

```python
# Warning: be careful when using complex matrices
C = np.array([ 4+1j, 3, 2-4j ])

print(C), print(' ')
print(C.T)
print(np.transpose(C))

# In python, we need to call Hermitian explicitly by first converting the complex conjugate and then flip
print(np.matrix(C).H)
```
